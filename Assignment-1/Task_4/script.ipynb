```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import multivariate_normal
```

```python
mean = [3, 4]
covariance = [[1, 0], [0, 2]]

x, y = np.mgrid[-10:10:.1, -10:10:.1]
pos = np.dstack((x, y))
z = multivariate_normal(mean, covariance)

# Plotting the 2D Gaussian distribution
plt.contourf(x, y, z.pdf(pos))
plt.title('2D Gaussian PDF plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.colorbar(label='Probability Density')
plt.show()
```

```python
def mahalanobisDistance(point, mean, cov):
    diff = (point - mean)
    term1 = (np.matmul(diff,cov)) 
    term2 = np.matmul(term1, diff)
    dist = np.sqrt(term2)
    return dist
```

```python
for point in [[10, 10], [0, 0], [3, 4], [6, 8]]:
    print(mahalanobisDistance(np.array(point), np.array(mean), np.array(covariance)))
```
